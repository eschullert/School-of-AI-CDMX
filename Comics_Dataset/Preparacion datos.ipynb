```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
dc=pd.read_csv('DC.csv')
marvel=pd.read_csv('Marvel.csv')
```

```python
dc['dc/marvel']='dc'
marvel['dc/marvel']='marvel'
marvel.rename(columns={'Year':'YEAR'}, inplace=True)
```

```python
df=pd.concat([dc,marvel],axis=0)
```

```python
df
```

```python
df.drop(columns=['urlslug'], inplace=True, errors='ignore')
```

```python
test=df.sample(frac=0.2)
```

```python
train=df.drop(index=test.index)
```

```python
len(train)+len(test)-len(df)
```

```python
train, test = train_test_split(df, test_size=0.25, random_state=42)
```

```python
train.set_index('page_id', inplace=True)
test.set_index('page_id', inplace=True)
```

```python
train.to_csv('train.csv')
test.to_csv('test.csv')
```
